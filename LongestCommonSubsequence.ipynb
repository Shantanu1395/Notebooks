**Choice Diagram**

               if x[-1] == y[-1]:
          /(yes)                   \(no)
     1+recurse on x-1,y-1        x + recurse on y-1,y + recurse on x-1

```python
def lcs(x,y,xLen,yLen):
    if xLen == 0 or yLen == 0:
        return 0
    else:
        if x[xLen-1] == y[yLen-1]:
            return 1+lcs(x,y,xLen-1,yLen-1)
        else:
            return max(lcs(x,y,xLen-1,yLen),lcs(x,y,xLen,yLen-1))

print(lcs("abcdgh","abedfhr",6,7))
```

```python
def lcs(x,y):
    result = [[0 for j in range(len(y)+1)] for i in range(len(x)+1)]
    for i in range(1, len(result)):
        for j in range(1, len(result[i])):
            if x[i-1] == y[j-1]:
                result[i][j] = 1 + result[i-1][j-1] 
            else:
                result[i][j] = max(result[i-1][j], result[i][j-1])

    for i in result:
        print(i)

    return result[-1][-1]

print(lcs("abcdgh","abedfhr"))
```

**2. Longest common substring**

---
break the continuity


```python
def lcs(x,y):
    result = [[0 for j in range(len(y)+1)] for i in range(len(x)+1)]
    ans = -1
    for i in range(1, len(result)):
        for j in range(1, len(result[i])):
            if x[i-1] == y[j-1]:
                result[i][j] = 1 + result[i-1][j-1] 
                ans = max(ans, result[i][j])    
            else:
                result[i][j] = 0

    for i in result:
        print(i)

    return ans

print(lcs("abcdgh","abcdfhr"))
```
