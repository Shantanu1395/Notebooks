**Problem** - Given and array and a number k, find if subset from an array sum =
 k

````python
arr = [3, 4, 5, 2]
k = 6
````

Similarity to Knapsack - 
1. Array is given
2. Choice is to be made whether to pick or not

            arr[i] < k
          /           \
         pick       can't pick 
(exclude or include)  


**Recursive Approach**

````python
def subsetSum(arr, k, n):
    if k == 0:
        return True
    if n == 0:
        return False  
    
    if arr[n-1] > k :
        return subsetSum(arr, k, n-1)
        
    return subsetSum(arr, k-arr[n-1], n-1) or subsetSum(arr, k, n-1)    
````

````python
print(subsetSum(arr, k, len(arr)))
````

````python
subsetSum = 11
dataset = [2, 3, 7, 8]
result = [[False for col in range(subsetSum+1)] for row in range(len(dataset)+1)]
    
#Base case
for i in result:
    i[0] = True

#Logic    
for i in range(1, len(result)):
    for j in range(1, len(result[0])):
        if j >= dataset[i-1]:
            result[i][j] = result[i-1][j-dataset[i-1]] or result[i-1][j]         
        else:
            result[i][j] = result[i-1][j]
for i in result:
    print(i)
````
