**1. Rod Cutting Problem - For each of the part, given price array, Cut the rod to have max profit**

---
input - price[], length

output - max profit

***Similarity to knapsack***

price[] -> val[]

length[] -> wt[]

len -> cap

---
Iterative

---


````python
price = [1, 5, 8, 9, 10, 17, 17, 20]
length = [1, 2, 3, 4, 5, 6, 7, 8]
l = len(price)

def cutRod(price, length, l):
    result = [[0 for j in range(l+1)] for i in range(len(price)+1)]
    
    for i in range(1, len(price)+1):
        for j in range(1, l+1):
            if length[i-1] <= j:
                result[i][j] = max(price[i-1] + result[i][j - length[i-1]], result[i-1][j])  
            else:      
                result[i][j] = result[i-1][j] 
    return result[len(price)][l]

print(cutRod(price, length, l))
````

**2. Coin change Problem - Maximum ways to compute sum with unlimited supply**

---
Similar to count of subset sum
---

input - coins[], sum

output - ways

````python
coins = [1, 2, 3]
s = 4

def coinChangeWays(coins, s, l):
    result = [[0 for j in range(s+1)] for i in range(len(coins)+1)]  

    for i in result:
      i[0] = 1  

    for i in range(1, len(coins)+1):
        for j in range(1, s+1):
            if coins[i-1]<=j:
                result[i][j] = result[i][j-coins[i-1]] + result[i-1][j]    
            else:
                result[i][j] = result[i-1][j]
    return result[len(coins)][s]

print(coinChangeWays(coins, s, 3))
````
